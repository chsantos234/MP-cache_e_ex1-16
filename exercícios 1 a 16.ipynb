exercício 2:
Quando a cache determina em qual conjunto certos blocos ficarão, pois esses blocos são previamente escolhidos a um conjunto estabelecido, e essa escolha é fixa, como no mapeamento direto, os blocos estão fixados a uma linha.

```python
def potenciaResultado(x): # retorna o expoente de um número inteiro
    multi = 1
    y = 0
    while multi < x:
        multi = multi * 2
        y = y + 1
    return y

print('ex 3')
# 1 bloco contem 8 células de 1 byte (LargLB = 8 bytes)
Qcelulas = 64 * pow(2,10)
largCelula = 1  # byte
LargLB = 8  # bytes
Qlinhas = 32

print('letra a:')

Qblocos = Qcelulas / LargLB
QblocosPorLinha = Qblocos / Qlinhas

Campotag = potenciaResultado(QblocosPorLinha)
print('largura da Tag:',Campotag,'bits')

CampoLinha = potenciaResultado(Qlinhas)
print('largura da linha:',CampoLinha,'bits')

CampoCelula = potenciaResultado(LargLB)
print('largura da célula:',CampoCelula,'bits')

END = Campotag + CampoLinha + CampoCelula
print('tamanho total do endereço:',END,'bits')

print('letra b:')

END = [0,0,0,1,0,0,0,1,0,0,0,1,1,0,1,1]
print('endereço:',*END, sep='')
print('Campo tag:',*END[:Campotag], sep='')
print('Campo linha:',*END[Campotag:Campotag + CampoLinha], sep='')
print('Campo endereço:',*END[Campotag + CampoLinha:], sep='')

print('letra c:')
CC = Qlinhas * LargLB
print('capacidade da cache:',CC,'bytes')
```
